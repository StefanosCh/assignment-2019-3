# Χαλασμένο Τηλέφωνο

Στο παιχνίδι του χαλασμένου τηλεφώνου, οιπαίχτες κάθονται σε μία σειρά. Ο πρώτος παίχτης επιλέγει μια λέξη και την ψυθιρίζει στον διπλανό του. Αυτός με τη σειρά του ψυθιρίζει τη λέξη που άκουσε στον διπλανό του, και πάει λέγοντας μέχρι τον τελευταίο παίχτη στη σειρά, ο οποίος αναφέρει στους υπόλοιπους τη λέξη που αυτός άκουσε. Συνήθως, η λέξη αυτή είναι πολύ διαφορετική από την αρχική λέξη που επέλεξε ο πρώτος παίχτης στην αρχή στης αλυσίδας. Σκοπός μας εδώ είναι να κάνουμε κάτι αντίστοιχο. 

Θέλουμε να φτιάξουμε ένα πρόγραμμα που να επιδεικνύει πώς μπορούν να αλλαχτούν λέξεις βηματικά, ώστε από μία αρχική λέξη, να προχωράμε σε νέες λέξεις, μέχρι μία τελική λέξη που έχουμε επιλέξει, με τον ελάχιστο αριθμό βημάτων. Επιπλέον, σε κάθε βήμα θέλουμε η μία λέξη να διαφέρει από την άλλη μόνο κατά ένα γράμμα:

* είτε θα σβήνουμε ένα γράμμα,

* είτε θα εισάγουμε ένα γράμμα,

* είτε θα αλλάζουμε ένα γράμμα.

Για παράδειγμα, έστω ότι θέλουμε να πάμε από τη λέξη spring στη λέξη summer. Η σειρά των μετατροπών θα είναι:

```
spring -> string -> sting -> ting -> tine -> time -> timer -> dimer -> dimmer -> simmer -> summer
```

Για να το κάνουμε αυτό, μπορούμε να σκεφτούμε ότι όλες οι λέξεις ενός λεξικού είναι οι κορυφές ενός γράφου. Δύο κορυφές (λέξεις) συνδέονται μεταξύ τους αν διαφέρουν κατά ένα γράμμα, με τον κανόνα που αναφέραμε παραπάνω. Τότε, το πρόβλημά μας ανάγεται στην αναζήτηση του συντομότερου μονοπατιού μεταξύ δύο λέξεων στο γράφο.

Στην πραγματικότητα όμως δεν χρειάζεται να κατασκευάσουμε ολόκληρο τον γράφο για να λύσουμε το πρόβλημά μας. Αρκεί να έχουμε έναν τρόπο ώστε από κάθε λέξη να μπορούμε να εντοπίσουμε τους γείτονές της, δηλαδή τις λέξεις που διαφέρουν από αυτήν κατά έναν χαρακτήρα.

### Απόσταση Levenshtein

Το πρώτο βήμα σε αυτήν την προσέγγιση είναι να έχουμε έναν τρόπο να μετράμε πόσο διαφέρουν δύο λέξεις, ώστε να μπορούμε να αποφαινόμαστε αν δύο λέξεις διαφέρουν κατά μία εισαγωγή, διαγραφή, ή αλλαγή γράμματος. Η μετρική η οποία μας δίνει ακριβώς τον αριθμό των εισαγωγών, διαγραφών, ή αλλαγών γραμμάτων που απαιτούνται για να μετατραπεί μία λέξη σε μία άλλη λέξη ονομάζεται [*απόσταση Levenshtein* (Levenshtein distance)](https://en.wikipedia.org/wiki/Levenshtein_distance).

Αυτή η μετρική απόστασης ορίζεται ως εξής. Αν έχουμε δύο συμβολοσειρές $a$ και $b$, όπου η $a$ έχει $i$ χαρακτήρες και η $b$ έχει $j$ χαρακτήρες, τότε:

* Αν είτε $i = 0$ είτε $j = 0$, τότε η μία συμβολοσειρά είναι άδεια, άρα η απόσταση θα είναι ίση με το μήκος της άλλης συμβολοσειράς.

* Διαφορετικά:

  1. Η $a$ μπορεί να είναι μεγαλύτερη από την $b$, οπότε υπολογίζουμε αναδρομικά την απόσταση μεταξύ των πρώτων $i - 1$ χαρακτήρων της $a$ και των $j$ χαρακτήρων της $b$ και προσθέτουμε ένα για τον χαρακτήρα που πρέπει να προσθέσουμε στη $b$.
  2. Η $b$ μπορεί να είναι μεγαλύτερη από την $a$, οπότε υπολογίζουμε αναδρομικά την απόσταση μεταξύ των πρώτων $j - 1$ χαρακτήρων της $b$ και των $i$ χαρακτήρων της $b$ και προσθέτουμε ένα για τον χαρακτήρα που πρέπει να προσθέσουμε στη $a$.
  3. H $a$ και η $b$ έχουν το ίδιο μήκος, οπότε υπολογίζουμε αναδρομικά την απόσταση μεταξύ των πρώτων $i - 1$ χαρακτήρων της $a$ και των $j - 1$ χαρακτήρων της $b$ και προσθέτουμε ένα αν χρειάζεται να αλλάξουμε το $a[i]$ (ή το $b[j]$).
  
  Η απόσταση είναι το καλύτερο από αυτά τα τρία ενδεχόμενα.

Αυτά περιγράφονται στον παρακάτω τύπο:

$$
d(a, i, b, j)= {
\begin{cases}
  \max(i,j) & {\text{ if }}\min(i,j)=0,\\
  \min {
     \begin{cases}
       d(a, i-1, b, j) + 1\\
       d(a, i, b, j-1) + 1\\
       d(a, i-1, b, j-1) + 1_{(a[i] \neq b[j])}
     \end{cases}} & {\text{ otherwise.}}
\end{cases}
}
$$

Ο συμβολισμός $1_{(a[i] \neq b[j])}$ σημαίνει 1 αν $a[i] \ne b[i]$ ή 0 αν  $a[i] = b[i]$.

Παρά το ότι μπορούμε να υλοποιήσουμε τον υπολογισμό της απόστασης Levenshtein ακολουθώντας απ' ευθείας τον αναδρομικό ορισμό, αυτό δεν είναι αποτελεσματικό, γιατί οι αναδρομικές κλήσεις υπολογίζουν ξανά και ξανά τις αποστάσεις μεταξύ προθεμάτων που έχουν ήδη υπολογίσει και στο παρελθόν. Στην πράξη, για να υπολογίσουμε την απόσταση Levenshtein χρησιμοποιούμε τον [αλγόριθμο των Wagner-Fisher](https://en.wikipedia.org/wiki/Wagner%E2%80%93Fischer_algorithm). Η ιδέα είναι ότι χρησιμοποιούμε έναν πίνακα για να αποθηκεύσουμε τις αποστάσεις μεταξύ όλων των προθεμάτων της πρώτης συμβολοσειράς και όλων των προθεμάτων της δεύτερης συμβολοσειράς και γεμίζουμε σταδιακά τον πίνακα.

### Δέντρα BK

Το δεύτερο βήμα για την επίλυση του προβλήματός μας είναι να βρούμε έναν αποτελεσματικό τρόπο για να εντοπίζουμε δυναμικά τους γείτονες κάθε λέξης. Ένας τέτοιος τρόπος μας δίνεται από μια δομή δεδομένων που ονομάζεται [δέντρο BK](https://en.wikipedia.org/wiki/BK-tree) (BK tree, από τα ονόματα των εφευρετών της δομής, Burkhard και Keller). Ένα δέντρο BK ορίζεται ως εξής: 

* Επιλέγουμε ένα οποιοδήποτε στοιχείο ως ρίζα.

* Κάθε κόμβος του δένδρου μπορεί να έκει οσαδήποτε παιδιά, έστω $n$, που αντιστοιχούν σε $n$ υποδέντρα. Το $k$ υποδέντρο ενός κόμβου περιέχει στοιχεία τα οποία έχουν απόσταση από τον κόμβο ίση με $k$. 

Ας δούμε πώς κατασκευάζεται ένα τέτοιο δέντρο· έτσι θα καταλάβουμε καλύτερα και τη μορφή του. 

1. Έστω ότι ξεκινάμε με τη λέξη food.

2. Εισάγουμε τη λέξη good. H απόσταση της good από food είναι 1, άρα θα η good θα γίνει παιδί της good, ενώ κάτω από το συγκεκριμένο κλαδί θα βάζουμε λέξεις που απέχουν 1 από τη food. Για να το δείξουμε αυτό, βάζουμε την ετικέτα 1 στο κλαδί.

3. Στη συνέχεια προσθέτουμε τη λέξη cook. Η απόσταση της cook από food είναι 2, άρα η cook θα γίνει παιδί της food μέσω άλλου κλαδιού όπου θα πηγαίνουν όλες οι λέξεις που απέχουν 2 από τη food. Βάζουμε την ετικέτα 2 στο νέο κλαδί.

4. Εισάγουμε τη λέξη fowl. Η απόσταση της fowl από τη food είναι 2, άρα προχωράμε στο κλαδί 2 της food, στον κόμβο cook. Εκεί δημιουργούμε νέο υποδέντρο. Η απόσταση της fowl από το cook είναι 3, άρα η fowl γίνεται παιδί της cook μέσω ενός νέου κλαδιού με ετικέτα 3.

5. Εισάγουμε τη λέξη spoon. Η απόσταση της spoon από food είναι 3, άρα η spoon θα γίνει παιδί της food μέσω νέου κλαδιού με ετικέτα 3 όπου θα πηγαίνουν όλες οι λέξεις που απέχουν 3 από τη food.

6. Εισάγουμε τη λέξη fork. Έχει απόσταση 2 από τη food, άρα πηγαίνουμε στον κόμβο cook. To fork έχει απόσταση 2 από το cook, άρα γίνεται παιδί της cook μέσω νέου κλαδιού με ετικέτα 2.

Τα δέντρα BK μας εξυπηρετούν διότι μας επιτρέπουν να βρούμε γρήγορα τις λέξεις οι οποίες έχουν απόσταση Levenshtein μέχρι ένα όριο $r$ από μία λέξη που μας ενδιαφέρει. Συγκεκριμένα:

1. Ξεκινάμε από τη ρίζα του δένδρου.

2. Έστω $D$ η απόσταση Levenshtein του τρέχοντα κόμβου από τη λέξη στόχο. Αν $D \le r$, τότε προσθέτουμε τη λέξη που είναι αποθηκευμένη στον τρέχοντα κόμβο στα αποτελέσματα.

3. Για κάθε παιδί του τρέχοντα κόμβου που το κλαδί του έχει ετικέτα $t$ με $D - r \le t \le D + r$ επιστρέφουμε στο βήμα 2 με το εν λόγω παιδί ως τρέχοντα κόμβο. 

Ο λόγος που ισχύει το παραπάνω οφείλεται στο ότι η απόσταση Levenshtein ακολουθεί την *τριγωνική ανισότητα*. Αν συμβολίσουμε με $d(x, y)$ την απόσταση μεταξύ δύο λέξεων $x$ και $y$, ισχύει για κάθε λέξη $w$:

$$d(x, y) \le d(x, w) + d(w, y)$$

Έστω ότι αναζητούμε τις λέξεις που απέχουν όχι περισσότερο από $r$ (δηλαδή έχουν ακτίνα, radius, $r$), από τη λέξη $w$. Έστω επίσης ότι βρισκόμαστε στον κόμβο $x$ του δένδρου και θέλουμε να δούμε αν θα πρέπει να εξετάσουμε το παιδί του $y$. Αφού δεν θέλουμε η $y$ να απέχει πάνω από $r$ από την $w$, έχουμε:

$$d(w, y) \le r$$

Τότε η τριγωνική ανισότητα γίνεται:

$$d(x, y) \le d(x, w) + r$$

Αν $D = d(x, w)$, παίρνουμε:

$$d(x, y) \le D + r$$

Μπορούμε όμως να χρησιμοποιήσουμε την τριγωνική ανισότητα και για την απόσταση $d(x, w)$:

$$d(x, w) \le d(x, y) + d(y, w)$$

Όπως και πριν, αφού δεν θέλουμε η $y$ να απέχει πάνω από $r$ από την $w$, έχουμε:

$$d(x, w) \le d(x, y) + r$$

ή 

$$d(x, y) \ge d(x, w) - r$$

Αντικαθιστώντας πάλι $D = d(x, w)$, παίρνουμε:

$$d(x, y) \ge D - r$$

Συνδυάζοντας τις δύο ανισότητες καταλήγουμε στο:

$$D - r \le d(x, y) \le D + r$$

### Εύρεση Μονοπατιού

Το τρίτο βήμα στην προσέγγιση είναι η αναζήτηση του μονοπατιού στο γράφο. Αυτό μπορεί να γίνει με μία τροποποίηση του αλγορίθμου του Dijkstra. Όπως και στον κανονικό αλγόριθμο, διατηρούμε μια ουρά προτεραιότητας την οποία χρησιμοποιούμε για να δούμε ποιον κόμβο θα επισκεφτούμε. Αντί όμως η προτεραιότητα των κόμβων στην ουρά να προκύπτει από την απόσταση από τον κόμβο που ξεκινήσαμε, θα προκύπτει από το άθροισμα της απόστασης από τον κόμβο που ξεκινήσαμε και την απόσταση μέχρι τον κόμβο προορισμού. Τότε στην πραγματικότητα παίρνουμε έναν άλλο αλγόριθμο αναζήτησης, τον [αλγόριθμο A*](https://en.wikipedia.org/wiki/A*_search_algorithm).

```
start_word, word_1, word_2, ..., target_word
```

Αν δεν μπορούμε να φτάσουμε από τη λέξη εκκίνησης στη λέξη προορισμού με τον τρόπο που έχουμε περιγράψει, η έξοδος του προγράμματος θα είναι απλώς η λέξη εκκίνησης:

```
start_word
```

Για λεξικό μπορείτε να χρησιμοποιήσετε το αρχείο [word_dict.txt](word_dict.txt). Το αρχείο έχει προκύψει από την επεξεργασία των λημμάτων του [WordNet](https://wordnet.princeton.edu/) με [αυτό το πρόγραμμα](wordnet_clean.py).

## Για Περισσότερες Πληροφορίες

Η μετρική Levenshtein πήρε το όνομά της από τον Σοβετικό μαθηματικό Vladimin Levenshtein, ο οποίος τη διερεύνηση το 1965. Βλ. Влади́мир И. Левенштейн (1965). Двоичные коды с исправлением выпадений, вставок и замещений символов. Доклады Академий Наук СCCP. 163 (4): 845–8. Στα αγγλικά η εργασία εκδόθηκε ως: Levenshtein, Vladimir I. (February 1966). "Binary codes capable of correcting deletions, insertions, and reversals". Soviet Physics Doklady. 10 (8): 707–710. 

Τα δέντρα BK εφευρέθηκαν από τους Walter Austin Burkhard και Robert M. Keller, βλ. W. A. Burkhard,  R. M. Keller. 1973. Some approaches to best-match file searching. Communications of the ACM 16, 4 (April 1973), 230-236. DOI=<http://dx.doi.org/10.1145/362003.362025>.
